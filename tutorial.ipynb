### Tutorial for MEDI Pipeline

1. Run initialize.py to get all functionally matched neurons' info from CAVE client.

```python
%run scripts/initialize.py
```

2. Run run_grating.py to get all neurons' tuning properties from drifting gratings.

```python
%run scripts/run_grating.py
```

3. Run run_generation.py to get all neurons' MEDI or MESI.
    
    You can terminate the process at any time. The next run will continue generating the unfinished parts.

```python
%run scripts/run_generation.py --mode=medi # default
%run scripts/run_generation.py --mode=mesi
```

4. Run run_medi_analysis.py to get all neurons' MEDI properties.

```python
%run scripts/run_medi_analysis.py
```

5. Run run_fit_gabor.py to get gabor fit evaluations, then run_gabor_generation.py to get optimal Gabor filters, then run_baseline_comparisons.py to get baseline responses, finally fig2.py and fig3.py to get Figure 2,3.

```python
%run scripts/run_fit_gabor.py
%run scripts/run_gabor_generation.py
%run scripts/run_baseline_comparisons.py
%run visualization/fig2.py
%run visualization/fig3.py
```

6. Run fig4.py ~ fig6.py to get Figure 4-6.

```python
%run visualization/fig4.py
%run visualization/fig5.py
%run visualization/fig6.py
```

7. Run run_pixel_generation.py to get example MEDI generated via optimization in the pixel space, then figs1.py to get Appendix Figure S1.

```python
%run scripts/run_pixel_generation.py
%run visualization/figs1.py
```

8. Run figs2.py ~ figs6.py to get Appendix Figure S2-6.

```python
%run visualization/figs2.py
%run visualization/figs3.py
%run visualization/figs4.py
%run visualization/figs5.py
%run visualization/figs6.py
```
